```python
# Dependencies
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
#Import CSV files
mouse_data_load = '/Users/wandabaghel/Documents/Classworkwk3/LearnPython/PythonMatplot Assignment/mouse_drug_data.csv'
clinical_data_load = '/Users/wandabaghel/Documents/Classworkwk3/LearnPython/PythonMatplot Assignment/clinicaltrial_data.csv'

# Load files
mousedrug_df = pd.read_csv(mouse_data_load)
clinicaltrial_df = pd.read_csv(clinical_data_load)

# Read the Mouse and Drug Data and the Clinical Trial Data
mousedrug_df.head()
```

```python
# Combine the data into a single dataset
clinic_mouse_df = pd.merge(clinicaltrial_df, mousedrug_df, on='Mouse ID')

# Display the data table for preview
clinic_mouse_df
```

```python
# Group by Drug and TimePoint
# Calculate mean of the Tumor Volume
# Store in Data Frame

tumor_mean_df = pd.DataFrame(clinic_mouse_df.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].mean())

# Display DataFrame
tumor_mean_df.head(20)

```

```python
# Group by Drug and TimePoint
# Calculate Standard error of the Tumor Volume
# Store in Data Frame

tumor_serror_df = pd.DataFrame(clinic_mouse_df.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].sem())

# Display DataFrame
tumor_serror_df.head(20)
```

```python
# Minor Data Munging to Re-Format the Data Frames

#tumor_pivot = tumor_mean_df.pivot
tumor_pivot_df = pd.pivot_table(tumor_mean_df, index='Timepoint', columns = 'Drug')

# Preview that Reformatting worked
tumor_pivot_df
```

```python
# Minor Data Munging to Re-Format the Standard Error Data Frames

serror_pivot_df = pd.pivot_table(tumor_serror_df, index='Timepoint', columns = 'Drug')

serror_pivot_df
```

```python
# Find Column Names
tumor_pivot_df.columns
```

```python
# Print Indexes
print(tumor_pivot_df['Tumor Volume (mm3)'].index)
```

```python
# 1. Generate the Tumor Response to Treatment Plot (with Error Bars)

x_axis = tumor_pivot_df['Tumor Volume (mm3)'].index
y_axis = tumor_pivot_df['Tumor Volume (mm3)']
s_error = serror_pivot_df['Tumor Volume (mm3)']

plt.figure(figsize=(8,5))

plt.errorbar(x_axis, y_axis['Capomulin'],yerr=s_error['Capomulin'], marker='o', markersize=5,
                label='Capomulin',color='r',linestyle='dashed',linewidth=0.5,capsize=3)
plt.errorbar(x_axis, y_axis['Infubinol'],yerr=s_error['Infubinol'], marker='^', markersize=5,
             label='Infubinol',color='b',linestyle='dashed',linewidth=0.5,capsize=3)
plt.errorbar(x_axis, y_axis['Ketapril'],yerr=s_error['Ketapril'], marker='s', markersize=5,
             label='Ketapril',color='g',linestyle='dashed',linewidth=0.5,capsize=3)
plt.errorbar(x_axis, y_axis['Placebo'],yerr=s_error['Placebo'], marker='d', markersize=5,
             label='Placebo',color='k',linestyle='dashed',linewidth=0.5,capsize=3)

plt.xlim(-5, 50)
plt.ylim(30, 75)
plt.grid(linestyle='dashed')
plt.xlabel('Time (Days)', fontsize=14)
plt.ylabel('Tumor Volume (mm3)', fontsize=14)
plt.title('Tumor Response to Treatment', fontsize=20)
plt.legend(loc='best', fontsize='small', fancybox=True)

# Save the figure
plt.savefig('TRestoTreat.png')

# Display Plot
plt.show()
```

```python
# 2. Metastaic Response to Treatment

# Group by Drug and TimePoint
# Calculate mean of the Metatstatic Sitess
# Store in Data Frame

meta_mean_df = pd.DataFrame(clinic_mouse_df.groupby(['Drug','Timepoint'])['Metastatic Sites'].mean())

#Display DataFrame
meta_mean_df.head(20)
```

```python
# Group by Drug and TimePoint
# Calculate Standard Error of the Metatstatic Sitess
# Store in Data Frame

meta_serr_df = pd.DataFrame(clinic_mouse_df.groupby(['Drug','Timepoint'])['Metastatic Sites'].sem())

#Display DataFrame
meta_serr_df.head(20)
```

```python
# Minor Data Munging to Re-Format the Mean Data Frames
meta_mean_df = meta_mean_df.reset_index()
meta_mean_pivot = meta_mean_df.pivot(index='Timepoint',columns='Drug')['Metastatic Sites']
meta_mean_pivot

```

```python
# Minor Data Munging to Re-Format the Standard Error Data Frames
meta_serr_df = meta_serr_df.reset_index()
meta_serr_pivot = meta_serr_df.pivot(index='Timepoint',columns='Drug')['Metastatic Sites']
meta_serr_pivot

```

```python
# Values for Drug Capomulin
meta_mean_pivot['Capomulin']
```

```python
# Generate the Metastatic Response to Treatment Plot (with Error Bars)

x_axis = meta_mean_pivot.index
y_axis = meta_mean_pivot
s_error = meta_serr_pivot

plt.figure(figsize=(8,5))

plt.errorbar(x_axis, y_axis['Capomulin'],yerr=s_error['Capomulin'], marker='o', markersize=5,
                label='Capomulin',color='r',linestyle='dashed',linewidth=0.5,capsize=0)
plt.errorbar(x_axis, y_axis['Infubinol'],yerr=s_error['Infubinol'], marker='^', markersize=5,
             label='Infubinol',color='b',linestyle='dashed',linewidth=0.5,capsize=0)
plt.errorbar(x_axis, y_axis['Ketapril'],yerr=s_error['Ketapril'], marker='s', markersize=5,
             label='Ketapril',color='g',linestyle='dashed',linewidth=0.5,capsize=0)
plt.errorbar(x_axis, y_axis['Placebo'],yerr=s_error['Placebo'], marker='d', markersize=5,
             label='Placebo',color='k',linestyle='dashed',linewidth=0.5,capsize=0)

plt.xlim(-5, 50)
plt.ylim(-0.5, 3.75)
plt.grid(linestyle='dashed')
plt.xlabel('Time (Days)', fontsize=14)
plt.ylabel('Metastatic Sites', fontsize=14)
plt.title('Metastatic Response to Treatment', fontsize=20)
plt.legend(loc='best', fontsize='small', fancybox=True)

# Save the figure
plt.savefig('MetaStaticResToTreatment.png')

# Display Plot
plt.show()
```

```python
# 3. Survival Rate

# Group by Drug and TimePoint
# Calculate count on Mouse ID
# Store in Data Frame

mouse_count_df = pd.DataFrame(clinic_mouse_df.groupby(['Drug','Timepoint'])['Mouse ID'].count())

mouse_count_df = mouse_count_df.rename(columns={'Mouse ID':'Mouse Count'})

mouse_count_df.head(20)
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_count_df = mouse_count_df.reset_index()

mouse_survive = mouse_count_df.pivot(index='Timepoint',columns='Drug')['Mouse Count']

mouse_survive
```

```python
# Survival Rate for the Capomulin drug.
survive_capo_percent = mouse_survive['Capomulin']/mouse_survive['Capomulin'].max()*100
survive_capo_percent
```

```python
# Generate the Survival During Treatment Plot (with Error Bars)

x_axis = mouse_survive.index
y_axis = mouse_survive

plt.figure(figsize=(8,5))

plt.errorbar(x_axis, y_axis['Capomulin']/y_axis['Capomulin'].max()*100, marker='o', markersize=5,
                label='Capomulin',color='r',linestyle='dashed',linewidth=0.5,capsize=0)
plt.errorbar(x_axis, y_axis['Infubinol']/y_axis['Infubinol'].max()*100, marker='^', markersize=5,
             label='Infubinol',color='b',linestyle='dashed',linewidth=0.5,capsize=0)
plt.errorbar(x_axis, y_axis['Ketapril']/y_axis['Ketapril'].max()*100, marker='s', markersize=5,
             label='Ketapril',color='g',linestyle='dashed',linewidth=0.5,capsize=0)
plt.errorbar(x_axis, y_axis['Placebo']/y_axis['Placebo'].max()*100, marker='d', markersize=5,
             label='Placebo',color='k',linestyle='dashed',linewidth=0.5,capsize=0)

plt.xlim(-5, 50)
plt.ylim(30, 105)
plt.grid(linestyle='dashed')
plt.xlabel('Time (Days)', fontsize=14)
plt.ylabel('Survival Rate (%)', fontsize=14)
plt.title('Survial During Treatment', fontsize=20)
plt.legend(loc='best', fontsize='small', fancybox=True)

# Save the figure
plt.savefig('SurvivalRate.png')

# Display Plot
plt.show()
```

```python
# 4. Summary Bar Graph

# Group by Drug and Timepoint. Calculate mean
tumor_df = clinic_mouse_df.groupby(['Drug','Timepoint'])
tumor_mid = tumor_df['Tumor Volume (mm3)'].mean()

# Reset the index
tumor_mid = tumor_mid.reset_index()

# Group by Drug 
tumor_mid_drug = tumor_mid.groupby('Drug')

# Calculate Percentage change between first and last Volume for each Drug
tum_md_dg_per = 100 * (tumor_mid_drug.last() - tumor_mid_drug.first()) / tumor_mid_drug.first()

# Reset the Index
tum_md_dg_per = tum_md_dg_per.reset_index()

# Drop Timepoint Column
tum_md_dg_per = tum_md_dg_per.drop(columns=['Timepoint'])

# Display the data to confirm
tum_md_dg_per
```

```python
# Summary Bar Plot
plt.figure(figsize=(8,5))

x_axis = np.arange(4)

plt.bar(tum_md_dg_per['Drug'][0], tum_md_dg_per['Tumor Volume (mm3)'][0],
        color='g',align='center',width=1,edgecolor='k',linestyle='--')
plt.bar(tum_md_dg_per['Drug'][2], tum_md_dg_per['Tumor Volume (mm3)'][2],
        color='r',align='center',width=1,edgecolor='k',linestyle='--')
plt.bar(tum_md_dg_per['Drug'][3], tum_md_dg_per['Tumor Volume (mm3)'][3],
        color='r',align='center',width=1,edgecolor='k',linestyle='--')
plt.bar(tum_md_dg_per['Drug'][5], tum_md_dg_per['Tumor Volume (mm3)'][5],
        color='r',align='center',width=1,edgecolor='k',linestyle='--')

plt.grid(True)
plt.xticks(x_axis,['Capomulin','Infubinol','Ketapril','Placebo'],fontsize=14)
plt.ylabel('% Tumor Volume Change',fontsize=14)
plt.title('Tumor Change Over 45 Day Treatment',fontsize=20)

#Display the plot
plt.show()
```

```python
# Mark plot with percentage change as label
plt.figure(figsize=(8,5))

# Create variable to pass in the text label function
rect1 = plt.bar(tum_md_dg_per['Drug'][0], tum_md_dg_per['Tumor Volume (mm3)'][0],
        color='g',align='edge',width=-1,edgecolor='k',linestyle='--')
rect2 = plt.bar(tum_md_dg_per['Drug'][2], tum_md_dg_per['Tumor Volume (mm3)'][2],
        color='r',align='edge',width=-1,edgecolor='k',linestyle='--')
rect3 = plt.bar(tum_md_dg_per['Drug'][3], tum_md_dg_per['Tumor Volume (mm3)'][3],
        color='r',align='edge',width=-1,edgecolor='k',linestyle='--')
rect4 = plt.bar(tum_md_dg_per['Drug'][5], tum_md_dg_per['Tumor Volume (mm3)'][5],
        color='r',align='edge',width=-1,edgecolor='k',linestyle='--')

# Function to Label text for Red Bars
def autolabelred(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., 3,
                '%d%%' % int(height),
                ha='center', va='bottom',color='white')

# Function to Label text for Green Bars        
def autolabelgreen(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., -10,
                '%d%%' % int(height),
                ha='center', va='bottom',color='white')

# Call the Red and Green Bar functions
autolabelgreen(rect1)
autolabelred(rect2)
autolabelred(rect3)
autolabelred(rect4)

# Plot the bar plot
plt.grid(True)
plt.xticks(x_axis,['Capomulin','Infubinol','Ketapril','Placebo'],fontsize=14)
plt.ylim(-30,70)
plt.ylabel('% Tumor Volume Change',fontsize=14)
plt.title('Tumor Change Over 45 Day Treatment',fontsize=20)

# Save the plot 
plt.savefig('TumorCngBar.png')

#Display the plot
plt.show()
```
